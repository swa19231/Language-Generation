NAIVE PROBABILISTIC METHOD OF GENERATING TEXT

**A short note about the code**:
                  I'm using bigrams and determining the frequenty occured word for any word in the corpus excluding the junk values. I save the frequently occured word for every word in a dictionary(nextword dictionary). I have a list of start words for generating the text. I also have a list of stop words but I'm not using that since the required text length is small (10). Using random I pick some word in the start word list and then use the function "sentence" to generate sentences of length 10.

```
import urllib
import nltk
import random
nltk.download('punkt')
corpus = [None for i in range(271)]
resource = urllib.request.urlopen('https://raw.githubusercontent.com/swa19231/Datasets/master/corpus.txt') 
trash=["g","m","km","m/s", "ms^-1","m/sec","km/hr","km/h",	"kmph",	"mi/hr",	"mi/h",	"mph",	"mi/s","mi/sec",
      "km/s","km/sec","m/s2","m/sÂ²",	"m/s^2",	"ms^-2","s",	"h",	"hr", '.']
start_words=[]
stop_words=[]
```

```
d={}
for i in range(271):
  corpus[i] = resource.readline().decode('utf8')
  for a in nltk.tokenize.sent_tokenize(corpus[i]):
    x=nltk.tokenize.word_tokenize(a)
    start_words.append(x[0])
    stop_words.append(x[-1])
    
    for j in range(len(x)-1):
      if (x[j],x[j+1]) not in d.keys() and (x[j+1],x[j]) not in d.keys():
        d[(x[j],x[j+1])]=1
      else:
        if (x[j],x[j+1]) not in d.keys():
          d[(x[j+1],x[j])]+=1
        else:d[(x[j],x[j+1])]+=1
    for k in range(len(x)):
      try:
        if x[k].isalpha()==False:
          trash.append(x[k])
      except:
        pass
  
```

```
d = dict([x for x in d.items() if x[0][0] not in trash and x[0][1] not in trash])
```

```
nextword={}
for (a,b) in d.keys():
  if a not in nextword.keys():
    nextword[a]=b
  else:
    if d[(a,b)]>d[(a,nextword[a])]:
      nextword[a]=b
```

```
start_words=list(set(start_words))
stop_words=set((stop_words))
```

```
def sentence(x,n):
  print(x)
  if n>=0:
    sentence(nextword[x],n-1)
    
```

Though the text generation generates the same set of words after some time that is because of common words like 'the', 'a', 'is' and so on the text becomes deterministic. This can be overcome to some extent by randomising betweeen top 3 frequently occuring words for a certain word.

```
nextword1={}
for (a,b) in d.keys():
  if a not in nextword1.keys():
    nextword1[a]=[b]
  else:
    if len(nextword1[a])<3:
      nextword1[a].append(b)
    else:
        index=0
        if d[(a,nextword1[a][0])]>d[(a,nextword1[a][1])]:
          index=1
          if d[(a,nextword1[a][1])]>d[(a,nextword1[a][2])]:
            index=2
        else:
          if d[(a,nextword1[a][0])]>d[(a,nextword1[a][2])]:
            index=2
        if d[(a,nextword1[a][index])]<d[(a,b)]:
          nextword1[a][index]=b
          
      
```

```
def sentence1(x,n):
  print(x)
  if n>=0:
    sentence1(random.choice(nextword1[x]),n-1)
    
```
